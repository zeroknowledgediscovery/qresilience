```python
import pandas as pd
from jupyterthemes import  jtplot
jtplot.style()
import pylab as plt
import numpy as np
```

```python
df1=pd.read_csv('./qmodels/taufemale.csv',sep=' ',header=None,names=['x','yr','att','tc','varc']).drop('x',axis=1)
df2=pd.read_csv('./qmodels/taumale.csv',sep=' ',header=None,names=['x','yr','att','tc','varc']).drop('x',axis=1)
df3=pd.read_csv('./qmodels/twhite.txt',sep=' ',header=None,names=['x','yr','att','tc','varc']).drop('x',axis=1)
df4=pd.read_csv('./qmodels/tblack.txt',sep=' ',header=None,names=['x','yr','att','tc','varc']).drop('x',axis=1)
df5=pd.read_csv('./qmodels/tau.csv',sep=',',header=None,names=['x','yr','tc','varc']).drop('x',axis=1).assign(att='all')
df6=pd.read_csv('./qmodels/tother.txt',sep=' ',header=None,names=['x','yr','att','tc','varc']).drop('x',axis=1)
df=pd.concat([df1,df2,df3,df4,df5,df6])
df=df.set_index('yr')
df=df[df.index>1973]
df['StdDev']=[0.5*np.sqrt(x) for x in df.varc.values]
```

```python
plt.style.use('fivethirtyeight')
plt.style.use('dark_background')

for lab in ['white','black','other']:
    df1=df[df.att==lab].ewm(alpha=.3).mean()
    df1=df1[df1.index>1982]
    df1=df1[df1.index<2023]

    df1.tc.plot(style='-',ms=3,label=lab)
    plt.fill_between(df1.index.values, df1['tc'] - df1['StdDev'], df1['tc'] + df1['StdDev'],  alpha=0.2)
    
plt.gca().set_title('time-constant for race')

#plt.gca().set_xlim(1980,2020)
plt.legend()
plt.savefig('race.png',dpi=300,bbox_inches='tight')
```

```python
plt.style.use('fivethirtyeight')
plt.style.use('dark_background')

for lab in ['male','female']:
    df1=df[df.att==lab].ewm(alpha=.3).mean()
    df1=df1[df1.index>1982]
    df1=df1[df1.index<2023]

    df1.tc.plot(style='-',ms=3,label=lab)
    plt.fill_between(df1.index.values, df1['tc'] - df1['StdDev'], df1['tc'] + df1['StdDev'],  alpha=0.2)
    
plt.gca().set_title('time-constant for sex')

#plt.gca().set_xlim(1980,2020)
plt.legend()
plt.savefig('sexes.png',dpi=300,bbox_inches='tight')
#plt.gca().set_xlim(1980,2020)
```

```python
import math
def triangle_area(a, b, c):
    """Calculate the area of a triangle given its side lengths using Heron's formula."""
    s = (a + b + c) / 2
    a=(s * (s - a) * (s - b) * (s - c))
    if a > 0:
         return math.sqrt(s * (s - a) * (s - b) * (s - c))
    else:
         return 0.

def calculate_changes(triangle1, triangle2):
    """Calculate changes in area and side lengths between two triangles."""
    area1 = triangle_area(*triangle1)
    area2 = triangle_area(*triangle2)

    area_change = area2 - area1
    #side_length_changes = [triangle2[i] - triangle1[i] for i in range(3)]

    return area_change
def getA(row):
    return triangle_area(row.R,row.L,row.RL)
```

```python
def resp(dpath):
    rw2008 = pd.read_csv(dpath,index_col=0).T.rename(columns={0:'R',1:'L',2:'RL'})
    rw2008['A'] = rw2008.apply(getA,axis=1)
    rw2008['dA'] = rw2008['A'].diff()
    rw2008=rw2008.fillna(0)
    return rw2008
```

```python
rw2008 = resp('./qmodels/relaxation_2004white.csv')
rb2008 = resp('./qmodels/relaxation_2004black.csv')
ro2008 = resp('./qmodels/relaxation_2004other.csv')

rw2008.A.plot(label='white')
rb2008.A.plot(label='black')
ro2008.A.plot(label='other')
plt.gca().legend()
plt.gca().set_ylim(0.005,None)
plt.gca().set_xlim(0,None)

plt.gca().set_title('recovery dynamics for race 2004')

#plt.gca().set_xlim(1980,2020)
plt.legend()
plt.savefig('recovery2004.png',dpi=300,bbox_inches='tight')
```
